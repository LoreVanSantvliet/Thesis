```python
tool_MGEs=phaster.MGE.drop_duplicates()

tool_mges_detected=[]
coverages=[]
dispersions=[]
for tool_MGE in tool_MGEs:
    #print(tool_MGE)
    detected_flag = False
    coverage_nr = 0
    dispersion_set=set()
    for row_1 in phaster.loc[phaster.MGE == tool_MGE].iterrows():
        gene = row_1[1][3]
        #print()
        #print()
        tool_contig = row_1[1][1] #phaster.contig.loc[phaster.MGE == tool_MGE][0]
        #print("contig: " + str(tool_contig))
        for self_row in mges.iterrows():
            #print(self_row[1][3])
            if (self_row[1][3] == gene) & (self_row[1][1] == tool_contig):
                detected_flag = True
                coverage_nr = coverage_nr + 1
                dispersion_set.add(mges.MGE[mges.gene_nr==gene].item())
    if(detected_flag):
        tool_mges_detected.append(tool_MGE)
        #print(tool_MGEs)
        coverages.append(coverage_nr / len(phaster.gene_nr.loc[phaster.MGE == tool_MGE]))
        dispersions.append(dispersion_set)
```

### General

```python
def get_performance_measures(tool_mges, self_mges):
    tool_MGEs=tool_mges.MGE.drop_duplicates()
    n_tool = len(tool_mges)
    tool_mges_detected=[]
    coverages=[]
    dispersions=[]
    for tool_MGE in tool_MGEs:
        detected_flag = False
        coverage_nr = 0
        dispersion_set=set()
        for gene in tool_mges.gene_nr.loc[tool_mges.MGE == tool_MGE]:
            if gene in self_mges.gene_nr.values:
                detected_flag = True
                coverage_nr = coverage_nr + 1
                dispersion_set.add(self_mges.MGE[self_mges.gene_nr==gene].item())
        if(detected_flag):
            tool_mges_detected.append(tool_MGE)
            coverages.append(coverage_nr / len(tool_mges.gene_nr.loc[tool_mges.MGE == tool_MGE]))
            dispersions.append(dispersion_set)
    print(tool_mges_detected)
    print(coverages)
    print(dispersions)
    average_coverage=np.mean(coverages)
    
    number_mges = []
    for mge_set in dispersions:
        number_mges.append(len(mge_set))
    average_dispersion=np.mean(number_mges)
    completeness = len(tool_mges_detected) / len(tool_MGEs)
    return [average_coverage, average_dispersion, completeness]
    
    
    
```
