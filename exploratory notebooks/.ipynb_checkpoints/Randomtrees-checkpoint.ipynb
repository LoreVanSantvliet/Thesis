# Random tree comparison, distribution

## Preparation

```python
import pandas as pd
from ete3 import PhyloTree
import random
import matplotlib.pyplot as plt
from pathlib import Path
import numpy as np
```

## Check distribution

```python
chars=["A", "B", "C", "D", "E", "F", "G", "H", "I", "J"]
```

```python
distances=[]
for i in range(10000):
    chars_r=random.sample(chars, len(chars))
    string="(((" + chars_r[0] + "," + chars_r[1] + "),((" + chars_r[2] + "," + chars_r[3] + ")," + chars_r[4] + ")),((" + chars_r[5] + "," + chars_r[6] + "),((" + chars_r[7] + "," + chars_r[8] + ")," + chars_r[9] + ")));"
    #string="\"(((" + chars_r[0] + ":0.00283," + chars_r[1] + ":0.00283):0.00283,((" + chars_r[2] + ":0.00283," + chars_r[3] + ":0.00283):0.00283," + chars_r[4] + ":0.00283):0.00283):0.00283,((" + chars_r[5] + ":0.00283," + chars_r[6] + ":0.00283):0.00283,((" + chars_r[7] + ":0.00283," + chars_r[8] + ":0.00283):0.00283," + chars_r[9] + ":0.00283):0.00283):0.00283):0.00283;\""
    tree1 = PhyloTree(string)
    #print(tree1)
    
    chars_r_2=random.sample(chars, len(chars))
    string_2="(((" + chars_r_2[0] + "," + chars_r_2[1] + "),((" + chars_r_2[2] + "," + chars_r_2[3] + ")," + chars_r_2[4] + ")),((" + chars_r_2[5] + "," + chars_r_2[6] + "),((" + chars_r_2[7] + "," + chars_r_2[8] + ")," + chars_r_2[9] + ")));"
    tree2 = PhyloTree(string_2)
    #print(tree2)
    
    comp=tree1.compare(tree2, has_duplications=True, unrooted=True)
    dist=comp.get("treeko_dist")
    
    distances.append(dist)
    
    
    
    
```

```python
print(distances)
```

```python
plt.hist(distances)
```
