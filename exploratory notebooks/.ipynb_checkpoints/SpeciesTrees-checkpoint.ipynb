```python
project_path=Path().resolve().parent
msa_path=project_path / "data"/ "supermatrix"
```

```python
output_path=project_path / "data" / "species_trees"
```

```python
pangenome=pd.read_csv(project_path / "data" / "filtered_pangenome.csv")
```

## Tree generator

```python
species_file="Apilactobacillus kunkeei_A.fasta"
species="Apilactobacillus kunkeei_A"
```

```python
def generate_speciestree(species):
    species_file=str(species+".fasta")
    aln = AlignIO.read(open(msa_path / species_file), 'fasta')
    calculator = DistanceCalculator('identity')
    dm = calculator.get_distance(aln)
    constructor = DistanceTreeConstructor()
    njtree = constructor.nj(dm)
    Phylo.write(njtree, output_path / str(species+".nex"), "nexus")
```
