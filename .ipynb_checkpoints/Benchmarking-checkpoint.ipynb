# Benchmarking "sophisticated" approach

## Preparation

```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
project_path = Path().resolve().parent
input_path = project_path / "results" / "intermediate" / "benchmarking" / "sophisticated"
#output_path = project_path / "results" / "MGE_files"
```

## CONJscan

```python
conjscan_soph = pd.read_csv(input_path / "conjscan_performance.csv")
conjscan_soph
```

```python
conjscan_soph.columns=['genome', 'score_threshold', 'n_threshold', 'fraction_threshold', 'n_scc_threshold', 'coverage', 'dispersion', 'recall', 'specificity', 'F-score']

```

## Phaster

```python
phaster_soph = pd.read_csv(input_path / "phaster_performance.csv")
phaster_soph
```

```python
phaster_soph.columns=['genome', 'score_threshold', 'n_threshold', 'fraction_threshold', 'n_scc_threshold', 'coverage', 'dispersion', 'recall', 'specificity', 'F-score']

```

## Joined for recall and precision

```python
joined_soph = pd.read_csv(input_path / "joined_performance.csv")
joined_soph
```

```python
joined_soph.columns = ['genome', 'score_threshold', 'n_threshold', 'fraction_threshold', 'n_scc_threshold', 'coverage', 'dispersion', 'recall', 'precision', 'F-score']

```

```python
joined_soph['method']='sophisticated'
joined_soph
```

```python
joined_soph.loc[:,['recall', 'coverage', 'dispersion']].boxplot(showmeans=True)
```

```python
joined_soph=joined_soph.pivot(index='genome', columns=['method'], values=['coverage', 'dispersion', 'recall', 'precision', 'F-score'])

joined_soph.columns.names=['measure', 'method']
joined_soph.columns=joined_soph.columns.reorder_levels(order=['method', 'measure'])
joined_soph
```

# Join three approaches

```python
simple_soph = pd.merge(joined, joined_soph, how="inner", on='genome', suffixes=("_simple", "_sophisticated"))
simple_soph.loc[:,['genome', 'coverage_simple', 'dispersion_simple', 'recall_simple', 'precision_simple', 'F-score_simple', 'coverage_sophisticated', 'dispersion_sophisticated', 'recall_sophisticated', 'precision_sophisticated', 'F-score_sophisticated']]
```

```python
joined_new=joined.pivot(index='genome', columns=['method'], values=['coverage', 'dispersion', 'recall', 'precision', 'F-score'])

joined_new.columns.names=['measure', 'method']
joined_new.columns=joined_new.columns.reorder_levels(order=['method', 'measure'])
joined_new
```
