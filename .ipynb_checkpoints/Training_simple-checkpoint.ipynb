# Training "simple" approach

## Preparation

```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
project_path = Path().resolve().parent
input_path = project_path / "results" / "intermediate" / "training" / "simple"
#output_path = project_path / "results" / "MGE_files"
```

## CONJscan

```python
conjscan = pd.read_csv(input_path / "conjscan_performance.csv")
conjscan
```

```python
conjscan.columns=['genome', 'n_threshold', 'coverage', 'dispersion', 'recall', 'specificity', 'F-score']

```

```python
df2=conjscan.pivot(index='genome', columns='n_threshold', values=['coverage', 'dispersion', 'recall', 'specificity', 'F-score'])
df2
```

```python
df2.loc[:,'coverage'].boxplot(showmeans=True)
```

```python
(np.isnan(df2.loc[:,'coverage']) == False)
```

```python
df2.loc[:,'dispersion'].boxplot(showmeans=True)
```

```python
df2.loc[:,'recall'].boxplot(showmeans=True)
```

```python
df2.loc[:, 'specificity'].boxplot(showmeans=True)
```
